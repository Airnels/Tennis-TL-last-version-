```python
def player_table(a_p,w_p):
    atp = a_p.copy()
    wta = w_p.copy()
    
    atp['player_name'] = atp['name_first'] + " " + atp['name_last']
    atp['dob'] = pd.to_datetime(atp['dob'],format='%Y%m%d',errors='coerce')
    atp['fed'] = 'ATP'
    atp = atp[['player_id','fed','player_name','ioc','dob','height','hand']]

    
    wta['player_name'] = wta['name_first'] + " " + wta['name_last']
    wta['dob'] = pd.to_datetime(wta['dob'],format='%Y%m%d',errors='coerce')
    wta['fed'] = 'WTA'
    wta = wta[['player_id','fed','player_name','ioc','dob','height','hand']]   
    
    player = pd.concat([atp,wta])
    player['player_id'] = player['player_id'].astype('category')
        
    return(player)
```

```python
def rank_table(a_rc,a_r,w_rc,w_r):
    atp1 = a_rc.copy()
    atp2 = a_r.copy()    
    atp1.rename(columns={'player':'player_id'},inplace=True)
    atp2.rename(columns={'player':'player_id'},inplace=True)
    atp1['fed'] = 'ATP'
    atp2['fed'] = 'ATP'
    atp = pd.concat([atp1,atp2])
    
    wta1 = w_rc.copy()
    wta2 = w_r.copy()
    wta1.rename(columns={'player':'player_id'},inplace=True)
    wta2.rename(columns={'player':'player_id'},inplace=True)
    wta1.drop(columns='tours',inplace=True)
    wta2.drop(columns='tours',inplace=True)
    wta1['fed'] = 'WTA'
    wta2['fed'] = 'WTA'
    wta = pd.concat([wta1,wta2])
        
    player_rc = pd.concat([atp1,wta1])
    player_r = pd.concat([atp2,wta2,player_rc])
    player23 = pd.merge(player,player_rc,how='inner',on=['player_id','fed'])
    player23['tl_23'] = [ 'I' if i in (I_atp23+I_wta23) else 'II' if i in (II_atp23+II_wta23) else 'III' if i in 
                         (III_atp23+III_wta23) else 'IV' if i in (IV_atp23+IV_wta23) else 'V' if i in (V_atp23+V_wta23) else
                         'VI' if j<200 else 'VII' for i,j in zip(player23['player_name'],player23['rank']) ]
    tl23 = player23[['player_id','tl_23']]
    player_r = pd.merge(player_r,tl23,on='player_id',how='outer')
    player_r = player_r[['player_id','fed','ranking_date','rank','points','tl_23']]
    
    return player_r
```

```python
def tourney_table(atp_m,atp_t,wta_m,wta_t):
    atp = atp_m.copy()
    atp_tour = atp_t.copy()
    wta = wta_m.copy()
    wta_tour = wta_t.copy()
    
    atp['fed'] = 'ATP'
    atp = atp.loc[atp['tourney_name'].isin(atp_tour)]
    atp = atp[['tourney_id','fed','tourney_name','tourney_level','surface','tourney_date']]
    atp['tourney_date'] = pd.to_datetime(atp['tourney_date'],format='%Y%m%d',errors='coerce')

    wta['fed'] = 'WTA'
    wta = wta.loc[wta['tourney_name'].isin(wta_tour)]
    wta = wta[['tourney_id','fed','tourney_name','tourney_level','surface','tourney_date']]
    wta['tourney_date'] = pd.to_datetime(wta['tourney_date'],format='%Y%m%d',errors='coerce')
    
    tourney = pd.concat([atp,wta])
    tourney.drop_duplicates(inplace=True)
    
    return(tourney)
```

# IV - MATCH TABLE

```python
def match_table(atp_m,atp_t,wta_m,wta_t):
    atp = atp_m.copy()
    atp_tour = atp_t.copy()
    wta = wta_m.copy()
    wta_tour = wta_t.copy()
    
    atp['fed'] = 'ATP'
    atp = atp.loc[atp['tourney_name'].isin(atp_tour)]

    wta['fed'] = 'WTA'
    wta = wta.loc[wta['tourney_name'].isin(wta_tour)]

    match = pd.concat([atp,wta])
    match = match[(match['score'].str.contains('RET')==False)&(match['score'].str.contains('W/O')==False)&
                 (match['score'].str.contains('Def')==False)]
    match = match[['tourney_id','fed','match_num','round','winner_id','winner_rank','loser_id','loser_rank','score','minutes']]
    
    return(match)
```
